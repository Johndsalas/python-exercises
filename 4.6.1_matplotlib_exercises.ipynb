```python
import matplotlib.pyplot as plt
import math as m

x = range(0,100)
y = [m.sqrt(n) for n in x]

plt.xlabel('Run')
plt.ylabel('Rise')

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import math as m

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Monster Mash')
ax1.plot(x, [m.sqrt(n) for n in x])
ax2.plot(x, [(n**3) for n in x], 'tab:orange')
ax3.plot(x, [m.tan(n) for n in x], 'tab:green')
ax4.plot(x, [2**n for n in x], 'tab:red')


```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.xlim(0, 20)
plt.ylim(-15, 20)

x = range(1, 20)
plt.title("Lines")

plt.xlabel('$Runs$')
plt.ylabel('$Rises$')

plt.plot( x, [m.sqrt(n) for n in x],label='$sqrt(x)$')
plt.plot( x, [(n**3) for n in x],label='$x^3$')
plt.plot( x, [m.tan(n) for n in x],label='$tan(x)$')
plt.plot( x, [2**n for n in x],label='$2^n$')
plt.legend()

```
